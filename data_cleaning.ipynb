## Geo Locations

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="MyApp")

def geolocate(country):
    try:
        loc = geolocator.geocode(country)
        return loc.latitude, loc.longitude
    except:
        return np.nan, np.nan
```

```python
clean_df
```

```python
countries = clean_df['country'].value_counts()
countries = countries.reset_index()
```

```python
countries
```

```python
countries['latitude'], countries['longitude'] = zip(*countries['index'].apply(geolocate))
countries.dropna(subset=['latitude', 'longitude'], how="any", inplace=True)
```

```python
countries.rename(columns={'index': 'country_name',
                    'country': 'news_num'}, inplace=True)
```

```python
countries
```

```python
countries.to_csv("data/geolocations.csv", index=False)
countries.to_csv("dashboard/data/geolocations.csv", index=False)
```
